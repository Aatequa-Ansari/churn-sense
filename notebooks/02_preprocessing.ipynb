# Data Preprocessing

This notebook focuses on preparing the customer churn dataset for machine learning.

Objectives:

- Remove irrelevant features
- Handle missing values
- Handle duplicate records
- Convert categorical variables into numerical form
- Prepare data for model training
- Prevent data leakage
- Create a clean machine learning dataset

````python
# Import required libraries

import pandas as pd
import numpy as np

# Load dataset

df = pd.read_csv(
    "../data/WA_Fn-UseC_-Telco-Customer-Churn.csv"
)

# Display shape

print("Dataset Shape:", df.shape)

# Display first rows

df.head()
````

# Dataset Information

Inspect dataset structure, data types, and missing values before preprocessing.

````python
# Display dataset information

df.info()
````

# Check Duplicate Records

Duplicate records can negatively affect model performance.

This step identifies duplicate observations within the dataset.

````python
# Count duplicate rows

duplicate_count = df.duplicated().sum()

# Display duplicate count

print("Duplicate Records:", duplicate_count)
````

# Check Unique Customer IDs

Customer ID should uniquely identify each customer.

This step verifies uniqueness before removing the identifier column.

````python
# Count unique customer IDs

unique_customers = df["customerID"].nunique()

# Total records

total_records = len(df)

# Display results

print("Unique Customer IDs:", unique_customers)

print("Total Records:", total_records)
````

# Remove Customer Identifier

Customer ID is a unique identifier and does not contain predictive information.

Keeping customerID may introduce noise and reduce model generalization.

Therefore, the column will be removed before model training.

````python
# Remove customerID column

df = df.drop(
    columns=["customerID"]
)

# Verify removal

df.head()
````

# Observation

- No duplicate records were found in the dataset.
- customerID contains unique values for every customer.
- customerID does not contribute useful information for churn prediction.
- The column was removed to improve model quality and prevent unnecessary complexity.

# TotalCharges Data Type Investigation

The TotalCharges column is currently stored as an object data type.

This is unexpected because TotalCharges represents monetary values and should be numerical.

The objective of this step is to identify invalid values and convert the column into a numeric format.

````python
# Check current data type

print(df["TotalCharges"].dtype)

# Display first values

df["TotalCharges"].head()
````

# Identify Invalid TotalCharges Values

Blank spaces inside the TotalCharges column prevent proper numerical conversion.

This step counts how many invalid records exist.

````python
# Count blank values

blank_count = (df["TotalCharges"] == " ").sum()

# Display result

print("Blank Values in TotalCharges:", blank_count)
````

# Convert TotalCharges to Numeric

Invalid values will be converted into missing values (NaN).

This allows numerical processing and statistical analysis.

````python
# Convert TotalCharges to numeric

df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)

# Verify data type

print(df["TotalCharges"].dtype)
````

# Missing Value Audit

After conversion, blank entries become missing values (NaN).

This step identifies the total number of missing values present in each column.

````python
# Check missing values

df.isnull().sum()
````

# Handle Missing Values

Since TotalCharges contains only a small number of missing values, median imputation will be used.

Median is robust against skewed distributions and outliers.

````python
# Fill missing values using median

df["TotalCharges"] = df["TotalCharges"].fillna(
    df["TotalCharges"].median()
)

# Verify missing values

df.isnull().sum()
````

# Observation

- TotalCharges was incorrectly stored as an object data type.
- Blank values were identified and converted to missing values.
- The column was successfully converted to a numerical format.
- Missing values were handled using median imputation.
- The dataset is now ready for feature encoding and machine learning preparation.

# Binary Feature Encoding

Binary categorical variables are converted into numerical format using label encoding.

This step prepares the data for machine learning algorithms.

````python
# Encode binary categorical features

binary_mapping = {
    "Yes": 1,
    "No": 0
}

# Encode gender

df["gender"] = df["gender"].map({
    "Female": 0,
    "Male": 1
})

# Encode Partner

df["Partner"] = df["Partner"].map(binary_mapping)

# Encode Dependents

df["Dependents"] = df["Dependents"].map(binary_mapping)

# Encode PhoneService

df["PhoneService"] = df["PhoneService"].map(binary_mapping)

# Encode PaperlessBilling

df["PaperlessBilling"] = df["PaperlessBilling"].map(binary_mapping)

# Encode target variable

df["Churn"] = df["Churn"].map(binary_mapping)

# Verify results

df.head()
````

# Observation

Binary categorical features were successfully converted into numerical format.

Encoding Summary:

- Female → 0
- Male → 1
- No → 0
- Yes → 1

The target variable (Churn) has also been converted into machine learning compatible format.

````python
# Verify encoded columns

df[
    [
        "gender",
        "Partner",
        "Dependents",
        "PhoneService",
        "PaperlessBilling",
        "Churn"
    ]
].head()
````

# One-Hot Encoding

Several categorical features contain more than two categories.

These variables will be transformed using One-Hot Encoding to make them suitable for machine learning algorithms.

This approach prevents the model from assuming an incorrect ordinal relationship between categories.

````python
# List multi-class categorical features

multiclass_features = [
    "MultipleLines",
    "InternetService",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies",
    "Contract",
    "PaymentMethod"
]

# Apply One-Hot Encoding

df = pd.get_dummies(
    df,
    columns=multiclass_features,
    drop_first=True
)

# Display shape after encoding

print("Dataset Shape:", df.shape)

# Display first rows

df.head()
````

# Observation

Multi-class categorical variables were transformed into numerical features using One-Hot Encoding.

Benefits:

- Prevents machine learning models from interpreting categories as ordered values.
- Retains important business information.
- Produces a fully numerical dataset suitable for model training.

The dataset is now significantly expanded because each category is represented by its own feature column.

# Feature Engineering

Feature Engineering creates new variables from existing data to improve predictive performance.

These engineered features capture customer value, loyalty, and subscription maturity, which are important indicators of churn behavior.

````python
# Create average monthly spend

df["AverageMonthlySpend"] = (
    df["TotalCharges"] /
    (df["tenure"] + 1)
)

# Create high value customer flag

df["HighValueCustomer"] = (
    df["MonthlyCharges"] >
    df["MonthlyCharges"].median()
).astype(int)

# Create long-term customer flag

df["LongTermCustomer"] = (
    df["tenure"] >= 24
).astype(int)

# Create tenure groups

df["TenureGroup"] = pd.cut(
    df["tenure"],
    bins=[-1, 12, 24, 48, 72],
    labels=[
        "New",
        "Growing",
        "Established",
        "Loyal"
    ]
)

# Display sample

df[
    [
        "tenure",
        "MonthlyCharges",
        "TotalCharges",
        "AverageMonthlySpend",
        "HighValueCustomer",
        "LongTermCustomer",
        "TenureGroup"
    ]
].head()
````

# Observation

New business-oriented features were created:

- AverageMonthlySpend estimates spending behavior across customer lifetime.
- HighValueCustomer identifies customers paying above the median monthly charge.
- LongTermCustomer identifies customers retained for at least 24 months.
- TenureGroup categorizes customers based on subscription age.

These engineered features can improve model performance and provide stronger business insights.

# Encode Tenure Group

````python
# One-hot encode tenure group

df = pd.get_dummies(
    df,
    columns=["TenureGroup"],
    drop_first=True,
    dtype=int
)

# Check shape

print(df.shape)

# Display first rows

df.head()
````

# Observation

TenureGroup was transformed into numerical features using One-Hot Encoding.
